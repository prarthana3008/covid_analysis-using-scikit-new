```python
missing_value_threshold = 0.5  # 50% threshold
missing_value_percent = df.isnull().mean()
missing_value_percent
columns_to_drop = missing_value_percent[missing_value_percent > missing_value_threshold].index

columns_to_drop

df_cleaned = df.drop(columns=columns_to_drop)
df_cleaned

# Filling missing values for numeric columns with median
numeric_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_cleaned[numeric_cols] = df_cleaned[numeric_cols].fillna(df_cleaned[numeric_cols].median())

# Converting date to datetime format
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'], format='%Y%m%d')
df
```
